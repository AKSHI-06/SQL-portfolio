```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats
```

```python
# Load the data
data = pd.read_csv('engagement_data.csv')
```

```python
# Calculate statistical metrics for days_diff_reg_watch
mean_reg_watch = data['days_diff_reg_watch'].mean()
median_reg_watch = data['days_diff_reg_watch'].median()
mode_reg_watch = data['days_diff_reg_watch'].mode()[0]  # Mode might return multiple values, we'll take the first one
```

```python
# Calculate statistical metrics for days_diff_watch_purch
mean_watch_purch = data['days_diff_watch_purch'].mean()
median_watch_purch = data['days_diff_watch_purch'].median()
mode_watch_purch = data['days_diff_watch_purch'].mode()[0]  # Mode might return multiple values, we'll take the first one
```

```python
# Print the results
print("Days Diff Reg-Watch:")
print(f"Mean: {mean_reg_watch}")
print(f"Median: {median_reg_watch}")
print(f"Mode: {mode_reg_watch}")
print("\nDays Diff Watch-Purch:")
print(f"Mean: {mean_watch_purch}")
print(f"Median: {median_watch_purch}")
print(f"Mode: {mode_watch_purch}")
```

```python
# Plot distributions
plt.figure(figsize=(14, 6))

# Distribution for date_diff_reg_watch
plt.subplot(1, 2, 1)
sns.histplot(data['days_diff_reg_watch'], kde=True, color='blue')
plt.axvline(mean_reg_watch, color='r', linestyle='--', label=f'Mean: {mean_reg_watch}')
plt.axvline(median_reg_watch, color='g', linestyle='-', label=f'Median: {median_reg_watch}')
plt.axvline(mode_reg_watch, color='b', linestyle='-', label=f'Mode: {mode_reg_watch}')
plt.title('Distribution of Days Diff Reg-Watch')
plt.legend()
# Distribution for date_diff_watch_purch
plt.subplot(1, 2, 2)
sns.histplot(data['days_diff_watch_purch'], kde=True, color='orange')
plt.axvline(mean_watch_purch, color='r', linestyle='--', label=f'Mean: {mean_watch_purch}')
plt.axvline(median_watch_purch, color='g', linestyle='-', label=f'Median: {median_watch_purch}')
plt.axvline(mode_watch_purch, color='b', linestyle='-', label=f'Mode: {mode_watch_purch}')
plt.title('Distribution of Days Diff Watch-Purch')
plt.legend()
plt.tight_layout()
plt.show()
```

## Analysis of Engagement and Conversion Metrics

### Days Diff Reg-Watch: the difference in days between the registration date and the date of first-time engagement

- **Mean**: `3.42` days
  - This indicates that on average, students engage with content about `3.42` days after registration.
- **Median**: `0.0` days
  - Most students engage immediately or very shortly after registration.
- **Mode**: `0` days
  - The most common engagement time is on the same day as registration.

### Days Diff Watch-Purch: the difference in days between the date of first-time engagement and the date of first-time purchase (NULL if they have no purchases)

- **Mean**: `26.25` days
  - On average, it takes `26.25` days for students to make a purchase after their first lecture.
- **Median**: `1.0` day
  - Most students convert to paying customers within `1` day of watching their first lecture.
- **Mode**: `0` days
  - Many students purchase on the same day as their first lecture viewing.
